```python
# Load library
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
# read dfs df that contain all tweets
df = pd.read_csv("dfs.csv", encoding="latin1")
```

```python
# make a list of the categories 
list1 = df.categories.tolist()
```

```python
# make an array of the recently created list
text_list = np.array(list1)
```

```python
# Create the bag of words feature matrix
count1 = CountVectorizer()
bag_of_words1 = count1.fit_transform(text_list)

# Show feature matrix
bag_of_words1.toarray()
```

```python
# Get feature names
feature_names1 = count1.get_feature_names()

# View feature names
feature_names1[0:10]
```

```python
# Create data frame
bag_df = pd.DataFrame(bag_of_words1.toarray(), columns=feature_names1)
```

```python
# describe the bag of words data frame
bag_df.describe()
```

```python
# import the necessary libraries for TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Create the tf-idf feature matrix
tfidf = TfidfVectorizer()
feature_matrix = tfidf.fit_transform(list1)

# Show tf-idf feature matrix
feature_matrix.toarray()
```

```python
# Show tf-idf feature matrix
#tfidf.get_feature_names()
```

```python
# Create data frame
tdif_df = pd.DataFrame(feature_matrix.toarray(), columns=tfidf.get_feature_names())
```

```python
# describe tfidf df all words
tdif_df.describe()
```

```python
# a summary of tdidf df
tdif_df.breweries.describe()
```

```python
# set df correlation greater than .20 
tdif_corr = tdif_df.corr()>0.20
```

```python
# inspect value counts for tfidf
categ = tdif_corr.venue.value_counts()
print(categ)
```

```python
# print when value is true
categ_true = tdif_corr.venue[tdif_corr.venue==True]
```

```python
# print first 10 values when tfidf is true
print(categ_true[0:10])
print(categ[0:10])
```
