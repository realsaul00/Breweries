```python
# Load library
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
df = pd.read_csv("dfs.csv", encoding="latin1")
```

```python
list1 = df.categories.tolist()
```

```python
text_list = np.array(list1)
```

```python
# Create the bag of words feature matrix
count1 = CountVectorizer()
bag_of_words1 = count1.fit_transform(text_list)

# Show feature matrix
bag_of_words1.toarray()
```

```python
# Get feature names
feature_names1 = count1.get_feature_names()

# View feature names
feature_names1
```

```python
# Create data frame
bag_df = pd.DataFrame(bag_of_words1.toarray(), columns=feature_names1)
```

```python
bag_df.describe()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create the tf-idf feature matrix
tfidf = TfidfVectorizer()
feature_matrix = tfidf.fit_transform(list1)

# Show tf-idf feature matrix
feature_matrix.toarray()
```

```python
# Show tf-idf feature matrix
tfidf.get_feature_names()
```

```python
# Create data frame
tdif_df = pd.DataFrame(feature_matrix.toarray(), columns=tfidf.get_feature_names())
```

```python
tdif_df.describe()
```

```python
tdif_df.breweries.describe()
```

```python
tdif_corr = tdif_df.corr()>0.20
```

```python
categ = tdif_corr.venue.value_counts()
print(categ)
```

```python
categ_true = tdif_corr.venue[tdif_corr.venue==True]
```

```python
print(categ_true)
print(categ)
```
